```python
#Adding columns in pyspark dataframe
from pyspark.sql import functions as f
df_pyspark=df_pyspark.withColumn('SexBinary', f.when( 
                     df_pyspark['Sex']=='male', 1).otherwise(0)
                     )
```

```python
df_pyspark.show(2)
```

```python
#doing Data transformation to find the present value of ticket price for each class
from datetime import datetime
df_pyspark=df_pyspark.withColumn('P_v_Fare',
                                 df_pyspark['Fare']*(1+0.07)**(
                                     datetime.now().year-1912)
                                )
```

```python
df_pyspark.toPandas().head(2)
```

```python
#dropping a column
df_pyspark.drop('P_v_Ticket').toPandas().head(2)
```

```python
df_pyspark.toPandas().head(5)
```

```python
#Renaming the columns
df_pyspark.withColumnRenamed('P_v_Fare', 'Inflation Adjusted Fare Price').columns
```

```python
df_pyspark.show(2)
```

```python
#Counting missing values per columns
missing_count={col: df_pyspark.filter(df_pyspark[col].isNull()).count() for col in df_pyspark.columns}
```

```python
for col,count in missing_count.items():
    print(f'{col} : {count}')
```

```python
# how==all. This will drop those rows where all values are null
missing_count={col: df_pyspark.na.drop(how='all').filter(df_pyspark[col].isNull()).count() for col in df_pyspark.columns}

for col,count in missing_count.items():
    print(f'{col} : {count}')
```

```python
# how==any with a threshold. This will drop those rows where atleast 2 values are null
missing_count={col: df_pyspark.na.drop(how='any',thresh=2).filter(df_pyspark[col].isNull()).count() for col in df_pyspark.columns}

for col,count in missing_count.items():
    print(f'{col} : {count}')
```

```python
#replacing the null values with the mean of the column
#For this we will be using imputive function

integer_cols= [col for col,dtype in df_pyspark.dtypes if dtype!="string"]

from pyspark.ml.feature import Imputer
imputer = Imputer(
        inputCols=integer_cols,
        outputCols = ["{}_imputed".format(c) for c in integer_cols]
).setStrategy("mean")
```

```python
imputer.fit(df_pyspark).transform(df_pyspark).toPandas().head(20)
```
